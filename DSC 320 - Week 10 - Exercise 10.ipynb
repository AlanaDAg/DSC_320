# <center> Calculating Probabilities from Data

# a.

```python
# Provide file path to data set for Python
file_path = r'C:\Users\alana\OneDrive\Desktop\DSC 320\week10data\survey_data.csv'
survey = pd.read_csv(file_path)

# Display data set on console
survey
```

```python
# Find the number of observations that have brown hair in the data frame
df_brown = survey[survey['hair_color'] == 'brown']
df_brown
```

```python
# Find the number of rows in the data frame
survey_rows = survey.shape[0]
# Find the number of rows that contain brown hair in the data frame
df_brown_rows = df_brown.shape[0]

# Find the probability that a college student has brown hair
brown_hair_prob = round((df_brown_rows / survey_rows) * 100, 2)
print(f"The probability that a college student has brown hair is: \n{brown_hair_prob}%")
```

# b.

```python
df_blue = survey[survey['eye_color'] == 'blue']
df_blue
```

```python
# Find the number of rows that contain blue eyes in the data frame
df_blue_rows = df_blue.shape[0]

# Find the probability that a college student has blue eyes
blue_eyes_prob = round((df_blue_rows / survey_rows) * 100, 2)
print(f"The probability that a college student has blue eyes is: \n{blue_eyes_prob}%")
```

# c.

```python
# Conditional probability
## Blue eyes GIVEN brown hair
### Find the probability
blue_given_brown = (survey[survey['hair_color'] == 'brown'] ['eye_color'] == 'blue').mean() * 100
blue_given_brown
print(f"The probability that a college student has blue eyes given they have brown hair is: \n{blue_given_brown:.2f}%")
```

# d.

```python
# Conditional probability
## Brown hair GIVEN blue eyes
### Find the probability
brown_given_blue = (survey[survey['eye_color'] == 'blue'] ['hair_color'] == 'brown').mean() * 100
brown_given_blue
print(f"The probability that a college student has brown hair given they have blue eyes is: \n{brown_given_blue:.2f}%")
```

# e.

> The above results indicate that no, these outcomes are dependent. Since these outcomes do not have the same probability, this indicates that they do not have the same chance of occurring.
>> According to this survey data, there is a higher probability that a college student who has blue eyes will also have brown hair. This survey data also shows that a student who has brown hair is less likely to also have blue eyes.
