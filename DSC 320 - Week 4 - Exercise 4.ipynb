# <center> Calculating Limits Numerically

# a.

```python
# Import necessary libraries
## NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return 2*x**3 - 4*x + 1

# Numerical calculation of the limit of f(x) as x approaches 3
x_values_a = np.linspace(2.5, 3.5, 100)
y_values_a = f(x_values_a)

# Plot the function
## Define visual attributes of graph
plt.plot(x_values_a, y_values_a, label='2x^3 - 4x + 1')
plt.axvline(x=3, color='magenta', label='x = 3 (limit)')
plt.title('Function f(x) and Limit Calculation')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

# b.

```python
# Define the function g(x)
def g(x):
    return np.exp(x - 1) / x

# Numerical calculation of the limit of g(x) as x approaches 0
x_values_b = np.linspace(-0.5, 0.5, 100)
y_values_b = g(x_values_b)

# Handling the division by zero for x = 0
y_values_b[np.isinf(y_values_b)] = np.nan

# Plot the function
## Define visual attributes of graph
plt.plot(x_values_b, y_values_b, label='e^(x-1) / x')
plt.axvline(x=0, color='fuchsia', linestyle='--', label='x = 0 (limit)')
plt.title('Function g(x) and Limit Calculation')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.legend()
plt.show()
```

# <center> Average Rate of Change

```python
# Define Average Rate of Change formula
def average_rate_of_change(f, a, b):
    return (f(b) - f(a)) / (b - a)

# Example use of function
f_example = lambda x: 3*x**2
a_example = 1
b_example = 2
average_rate_of_change(f_example, a_example, b_example)
```
