# <center> Riemann Sum Function

```python
# Import necessary libraries
 ## NumPy for mathematical operations on arrays; matplotlib.pyplot to create graphs
import numpy as np
import matplotlib.pyplot as plt

# Define Right Riemann Sum function as a Python function
def Right_Riemann_Sum(f, a, b, n):
    
    delta_x = (b-a) / n
    x_right = np.linspace(delta_x, b, n)
    right_riemann_sum_output = np.sum(f(x_right) * delta_x)
    
    return right_riemann_sum_output
    # Use 'return' so Python knows it is performing a calculation
```

```python
# Test the Right_Riemann_Sum function
 ## Pass values to test variables

f_tst = lambda x:1 / (1+x**2)
a_tst = 0
b_tst = 5
n_tst = 10

test_right_riemann_sum = Right_Riemann_Sum(f_tst, a_tst, b_tst, n_tst)
print(test_right_riemann_sum)
```

# <center> Using the Right Riemann Sum to Find the Area Under a Curve

# a.

```python
# Import necessary libraries
 ##  ## NumPy for mathematical operations on arrays; matplotlib.pyplot to create graphs
import numpy as np
import matplotlib.pyplot as plt

# Define given function in Python f(x)=sqrt(4-x^2)
 ## Use np.sqrt() function for square root
f = lambda x: np.sqrt(4 - x**2)

# Define interval [0,2] = [a,b]
a = 0
b = 2

# Define values of n, x, and y
 ## np.linspace() function creates evenly spaced values inside the defined interval
n = 4
x = np.linspace(a, b, n+1)
y = f(x)

# Define the plotted x and y values
x_right = x[1:]
y_right = y[1:]

# Create graph of function with Right Riemann Sum rectangles
 ## Provide plotted variables and define aesthetics
plt.plot(x_right, y_right, 'm', markersize = 10)
plt.bar(x_right, y_right, width=-(b-a) / n, alpha=0.3, align='edge', linewidth=1.5, edgecolor='k')
plt.title('Sketch of $f(x) = \sqrt{4 - x^2}$ and Right Riemann Sum, where n=4')
plt.grid(True)
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.show()

# NOTE: The fourth rectangle is not visible because it extends from 1.50 to 2.00, so it sits along the x-axis.
```

# b.

```python
# Calculate the Right Riemann sum of R where n=4
 ## Variables and function are defined from part 1.

# Define Right Riemann Sum function as a Python function
def Right_Riemann_Sum(f, a, b, n):
    
    delta_x = (b-a) / n
    x_right = np.linspace(delta_x, b, n)
    right_riemann_sum_output = np.sum(f(x_right) * delta_x)
    
    return right_riemann_sum_output
    # Use 'return' so Python knows it is performing a calculation
    
# Calculate and display Right Riemann Sum
part2_right_riemann_sum = Right_Riemann_Sum(f, a, b, n)
print(part2_right_riemann_sum)
```

# c.

```python
# Incrementally increase the value of n to find the definite integral aka the area below the curve

for n in [100, 1000, 5000, 50000, 19900000]:
    part2_right_riemann_sum = Right_Riemann_Sum(f, a, b, n)
    print(f" Right Riemann Sum when n={n}:", part2_right_riemann_sum)
```
