# 1. Data Normalization

```python
# No need to import package here; min() and max() are already inside Python

# Create user defined function for data normalization
## Inputs an array with [] (for lists)
def Normalization_data(data_list):
    results = []
    # Assign objects for variables in formula
    min_value = min(data_list)
    max_value = max(data_list)
    denominator = max_value - min_value
    
    # Use 'for' loop because we are working with a list
    ## Create the actual calculation for normalization
    for x in data_list:
        new_x = (x - min_value) / denominator
        results.append(new_x)
        
    # Use 'return' so Python knows you are performing a calculation
    return results
```

```python
# Test no.1 function

test_list = [2, 4, 7, 9, 12, 15, 26, 39, 70]

test_results = Normalization_data(test_list)
print(test_results)
```

# 2. Data Standardization

```python
# NOTES:
## z-scores (statistics): mean divided by the standard deviation
## x-bar is sample mean
## s is standard deviation

# Import necessary libraries
## 'Statistics' to perform statistical calculations (standard deviation)
## NumPy for working with arrays
import statistics
import numpy as np

# Create user defined function for data standardization
## Assign objects to variables in formula
def Standardized_data(listed_data):
    new_results = []
    x_bar = statistics.mean(listed_data)
    stnd_dev = np.std(listed_data)
    
    # 'for' loop again
    ## Write the calculation for standardization
    for x in listed_data:
        new_x = (x - x_bar) / stnd_dev
        new_results.append(new_x)
        
    # 'return' keyword again    
    return new_results
```

```python
# Test no.2 function

list_for_test = [5.1, 7.2, 9.3, 12.4, 15.5, 89, 60, 45, 84, 23, 26, 36]

test_2_results = Standardized_data(list_for_test)
print(test_2_results)
```
