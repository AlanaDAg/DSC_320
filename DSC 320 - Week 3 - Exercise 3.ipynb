```python
# Import libraries
## Pandas to perform data manipulation
## Matplotlib to create scatter plot
## Scikit-learn linear model & LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Provide dataset file location for Python
file_path = r'C:\Users\alana\OneDrive\Desktop\DSC 320\us_pop_data.csv'
data = pd.read_csv(file_path)

# Convert table to dataframe
df = pd.DataFrame(data)

# Insert two new columns (years since 1790 & population in millions)
df.insert(2, 'yrs_since_1790', df['year'] - 1790)
df.insert(3,'pop_in_millions', df['us_pop'] / 1000000)

# Data prep
x = df[['yrs_since_1790']]
y = df['pop_in_millions']

# Create linear regression model
model = LinearRegression()
model.fit(x,y)

# Calculate the slope (coefficient) and y-intercept of the best-fit line
slope_best_fit = model.coef_[0]
intercept_best_fit = model.intercept_

# Create scatter plot (US population in millions VS years since 1790)
## Define a scatter plot point color
line_color = 'lime'
point_color = 'maroon'

# Define plot type using Matplotlib function plt.scatter(); include parameters
plt.scatter(df['yrs_since_1790'], df['pop_in_millions'], color = point_color, label = "Population Growth")

# Create linear regression model
plt.plot(df['yrs_since_1790'], slope_best_fit * data['yrs_since_1790'] + intercept_best_fit, color = 'darkorange', label = 'Linear Regression')

# Define visual attributes of graph
plt.title('Population in Millions vs Years Since 1790')
plt.xlabel('Years since 1790')
plt.ylabel('Population in Millions')
plt.legend()
plt.show()

# Define prediction method
y_pred = model.predict(x)

# Calculate R2 value and report on console
r2 = r2_score(y, y_pred)
print(f'R2 = {r2}')
```

```python
# Import necessary libraries
## Pandas, NumPy, and Matplotlib (.pyplot)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Provide dataset file location for Python
customer_spending_data = pd.read_csv(r'C:\Users\alana\OneDrive\Desktop\DSC 320\customer_spending.csv')

# Create a new column with log-transformed annual spending
customer_spending_data['log_ann_spending'] = np.log1p(customer_spending_data['ann_spending'])

# Create a histogram of transformed annual spending
## Define aesthetics
plt.figure(figsize=(10, 6))
plt.hist(customer_spending_data['log_ann_spending'], bins=20, edgecolor='black', color='orangered')
plt.title('Number of Customers vs Annual Spending of Transformed Data')
plt.xlabel('Logarithm Transformed Annual Spending')
plt.ylabel('Number of Customers')
plt.show()
```

# d.

> The histogram of the initial data appears to be skewed to the left while the version made with log transformed data is much more evenly distributed. This provides a great example of why it is useful to apply log transformations to data for modeling purposes. Modeling of the initial data can provide skewed results, which could lead to inaccurate ideas and decisions made based on that visual. It is important to be able to identify outliers but it could be difficult to understand what is happening to the majority of the data if only using the initial histogram.
