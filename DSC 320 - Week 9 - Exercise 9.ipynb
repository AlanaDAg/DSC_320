# <center> Finding the Eigenvalues and Eigenvectors of a Matrix

# a.

```python
# Input matrix A into Python by assigning it as an array

# Import libraries
import numpy as np

A = np.array([
    [4, 0, 1],
    [-1, -6, -2],
    [5, 0, 0]
])

print(A)
# Confirm shape of matrix A
A.shape
```

```python
# Assign vector as an array
v1 = np.array([1, 2, 3])
# Confirm shape of array
print(v1.shape)

# Determine if vector is an eigenvector of matrix A
## Do so by multiplying using np.dot() function
print(np.dot(A, v1))


# Multiply vector by 2 and compare with above calculation
q1a = v1 * 2
# Confirm shape of array
print(q1a.shape)

print(q1a)
```

```python
# Try confirming in another way
## Using np.array_equal() function

np.array_equal(A*v1, 2*v1)
```

> No, the vector [1, 2, 3] is not an eigenvector of matrix A.

# b.

```python
# Assign vector as an array
v2 = np.array([0, 1, 0])
# Confirm shape of array
print(v2.shape)

# Determine if vector is an eigenvector of matrix A
## Do so by multiplying using np.dot() function
print(np.dot(A, v2))


# Multiply vector by 1 and compare with above calculation
q1b = v2 * 1
# Confirm shape of array
print(q1a.shape)

print(q1b)
```

```python
# Try confirming in another way
## Using np.array_equal() function

np.array_equal(A*v2, 1*v2)
```

> No, the vector [0, 1, 0] is not an eigenvector of matrix A.

# c.

```python
# Use np.linalg.eig(matrix) from NumPy

# Import library and package
from numpy import linalg as LA
from numpy.linalg import eig

eigenvalues, eigenvectors = LA.eig(A)
print(f"The eigenvalues of matrix A are: \n{eigenvalues}")
print(f"The eigenvectors of matrix A are: \n{eigenvectors}")
```

# <center> Principal Component Analysis

# a.

```python
# Import library
## Pandas for data manipulation
import pandas as pd

# Provide dataset file path for Python
file_path = r'C:\Users\alana\OneDrive\Desktop\DSC 320\video_game_data.csv'
df_vgames = pd.read_csv(file_path)

# Display dataframe on console
df_vgames
```

```python
# Make a scatterplot of the dataset
## Import libraries
import matplotlib.pyplot as plt

plt.scatter(df_vgames['critic_score'], df_vgames['user_score'], 
           c='fuchsia')
# Define scatterplot details
plt.title("Video Game User Scores vs Critic Scores")
plt.ylabel("User Scores")
plt.xlabel("Critic Scores")
plt.show()
```

# b.

```python
# Calculate the covariance matrix first
cov = np.cov(df_vgames[['critic_score', 'user_score']].T)

# Determine eigenvalues/eigenvectors of the covariant matrix
eigenvalues, eigenvectors = LA.eig(cov)

# Sort eigenvectors 
sort_eigvals = np.argsort(eigenvalues)[::-1]
sorted_eigenvectors = eigenvectors[:, sort_eigvals]

# Original scatterplot of video game data frame
plt.scatter(df_vgames['critic_score'], df_vgames['user_score'])

# Use plt.quiver() function to create arrows for the Principal Components
## Principal Components (aka the eigenvectors of the covariance matrix)
### First Principal Component
PC1 = plt.quiver(*df_vgames[['critic_score', 'user_score']].mean(),
           *sorted_eigenvectors[:, 0] * np.sqrt(eigenvalues[0]),
           color='fuchsia', scale=1, scale_units = 'xy', angles = 'xy')

### Second Principal Component
PC2 = plt.quiver(*df_vgames[['critic_score', 'user_score']].mean(),
           *sorted_eigenvectors[:, 1] * np.sqrt(eigenvalues[1]),
           color='springgreen', scale=1, scale_units='xy', angles='xy')

# Define aesthetics and scatterplot details
plt.title("Video Game User Scores vs Critic Scores")
plt.ylabel("User Scores")
plt.xlabel("Critic Scores")
# Quiverkey labels
plt.quiverkey(PC1, 8.4, 9.4, 2, 'PC1', coordinates='data', alpha=0)
plt.quiverkey(PC2, 8.6, 8.6, 2, 'PC2', coordinates='data', alpha=0)
plt.grid(True)
plt.show()
```
