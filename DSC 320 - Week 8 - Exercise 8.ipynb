# <center> Solving a System of Linear Equations

```python
# Import Libraries
import numpy as np

# Convert system of linear equations to a matrix
## Define A matrix as an object
### Use np.array() function (N-dimensional)
A = np.array((
    (27, -10, 4, 29),
    (-16, 5, -2, 18),
    (-17, 4, -2, 20),
    (-7, 2, -1, 8)
))

print(A)
```

```python
# Assign right-hand values as a matrix object
righthand = np.matrix(((1),(-1),(0),(1)))
print(righthand)

# See shape of matrix b
righthand.shape
```

```python
# Transpose right-hand values
## Change shape from 1 row, 4 columns to 4 rows, 1 column
righthand_1 = np.transpose(righthand)
print(righthand_1)
```

```python
# Confirm shape of transposed b matrix
righthand_1.shape
```

```python
# Use linear algebra package from Numpy library to invert matrix A
## Import linear algebra package
from numpy.linalg import inv

A_inv = inv(A)
print(A_inv)
```

```python
# Use np.dot() function to multiply inverted matrix A and original A
## This shows the identity matrix
np.dot(A_inv, A)
```

```python
# Ax = b is the same as x = A^-1 * b
## Find value of x by multiplying inverted matrix A with right-hand values
### Again, use np.dot() function

x = np.dot(A_inv, righthand_1)
print(x)
```

```python
# Check that above answer (value of x) is true

## Multiply original A matrix with value of x
Check_A = np.dot(A, x)
print(Check_A)

# Remember right-hand values of original A matrix
print(righthand_1)

# The values are extremely close to the right-hand values in the original A matrix
## Note: -6.66133815e-16 is extremely close to 0
```

```python
# Assign square root of 2 divided by 2 as an object
## This will make it easier to work with rather than using math.sqrt() function
### NOTE: 2^0.5 is equal to the square root of 2

SQRT2_over2 = 2**0.5 / 2

# In Python, recreate matrix A using newly assigned object
A = np.matrix((
    (SQRT2_over2, -SQRT2_over2),
    (SQRT2_over2, SQRT2_over2)
))

print(A)
```

```python
#
x = np.matrix((
    (2),
    (-2)
))

# View v values
## See shape of v value matrix
print(x)
x.shape
```

```python
# Transpose v value matrix
## Change shape from 1 row, 2 columns to 2 rows, 1 column

x_1 = np.transpose(x)
print(x_1)
```

```python
# Confirm correct shape of v value matrix
x_1.shape
```

```python
# Find T(v)

T_x = np.dot(A, x_1)
print(T_x)
```

```python
# From section 7.1.3 from Essential Math for Data Science

# Import libraries
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
print(x)
print(y)

xx, yy = np.meshgrid(x, y)

plt.scatter(xx, yy, s=20, c=xx+yy)
# [. . .] Add axis, x and y with the same scale

plt.axvline(0, c='black', ls='-')
plt.axhline(0, c='black', ls='-')
plt.show()

T = np.array([
    [-1, 0],
    [0, -1]
])

xy = np.vstack([xx.flatten(), yy.flatten()])
xy.shape

T @ xy[:, 0]

trans = T @ xy
trans.shape

xx_transformed = trans[0].reshape(xx.shape)
yy_transformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots (1, 2, figsize=(6, 3))
axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
# [. . .] Add axis, x and y with the same scale

axes[0].axvline(0, c='black', ls='-')
axes[0].axhline(0, c='black', ls='-')
axes[1].axvline(0, c='black', ls='-')
axes[1].axhline(0, c='black', ls='-')


T = np.array([
    [1.3, -2.4],
    [0.1, 2]
])
trans = T @ xy

xx_transformed = trans[0].reshape(xx.shape)
yy_tranformed = trans[1].reshape(yy.shape)

f, axes = plt.subplots(1, 2, figsize=(6, 3))

axes[0].scatter(xx, yy, s=10, c=xx+yy)
axes[1].scatter(xx_transformed, yy_transformed, s=10, c=xx+yy)
# [...] Add axis, x and y with the same scale

# Add axis lines
axes[0].axvline(0, c='black', ls='-')
axes[0].axhline(0, c='black', ls='-')
axes[1].axvline(0, c='black', ls='-')
axes[1].axhline(0, c='black', ls='-')
```
