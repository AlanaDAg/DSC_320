```python
# Import libraries
## matplotlib.pytplot to create scatterplot
## Pandas to perform data manipulation
import matplotlib.pyplot as plt
import pandas as pd

# Povide dataset file location for Python
file_path = r'C:\Users\alana\OneDrive\Desktop\DSC 320\car_data.csv'
data = pd.read_csv(file_path)

# Define a scatter plot point color
point_color = 'mediumorchid'

# Use interpolation to "eyeball" the line of best-fit
slope_estimated = -0.01
intercept_estimated = 56

# Define visual attributes of scatter plot
## Display scatter plot on console
plt.scatter(data['weight'], data['hwy_mpg'], color = point_color, label = 'Provided Values')
plt.title('Highway MPG vs Weight')
plt.xlabel('Weight (pounds)')
plt.ylabel('Highway MPG')

# Plot the estimated line of best-fit
## Display scatter plot on console
plt.plot(data['weight'], slope_estimated * data['weight'] + intercept_estimated, color = 'midnightblue', label =' Estimated Best-Fit \n (Interpolation)')
plt.legend()
plt.show()
```

# 6.

```python
# Import libraries
## matplotlib.pytplot to create scatterplot
## Pandas to perform data manipulation
## Scikit-learn linear model & LinearRegression
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

# Povide dataset file location for Python
file_path = r'C:\Users\alana\OneDrive\Desktop\DSC 320\car_data.csv'
data = pd.read_csv(file_path)

# Define a scatter plot point color
point_color = 'mediumorchid'

# Data preparation
## Set x and y values based on column labels
X = data[['weight']]
y = data['hwy_mpg']

# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Calculate the slope (coefficient) and y-intercept of the best-fit line
slope_best_fit = model.coef_[0]
intercept_best_fit = model.intercept_

# Plot the scatterplot
plt.scatter(data['weight'], data['hwy_mpg'], color = point_color, label = 'Provided Values')

# Plot the eyeballed fitted line
plt.plot(data['weight'], slope_estimated * data['weight'] + intercept_estimated, color = 'midnightblue', label = 'Estimated Best-Fit \n (Interpolation)')

# Plot the best-fit line
# Define visual attributes of scatter plot
## Display scatter plot on console
plt.plot(data['weight'], slope_best_fit * data['weight'] + intercept_best_fit, color = 'darkorange', label = 'Actual Best-Fit Line')
plt.title('Highway MPG vs Weight')
plt.xlabel('Weight (pounds)')
plt.ylabel('Highway MPG')
plt.legend()
plt.show()
```

# 7.

```python
# Import libraries
## Scikit-learn to calculate MSE
## NumPy to work with arrays
from sklearn.metrics import mean_squared_error
import numpy as np

# Predictions for the estimated line
y_pred_manual = slope_estimated * data['weight'] + intercept_estimated

# Predictions for the best-fit line
y_pred_best_fit = model.predict(X)

# Calculate the RMSE for the predicted and actual lines of best-fit
rmse_manual = np.sqrt(mean_squared_error(y, y_pred_manual))
rmse_best_fit = np.sqrt(mean_squared_error(y, y_pred_best_fit))

print(f"The RMSE for the estimated best-fit line is: {rmse_manual:.2f}")
print(f"The RMSE for the actual best-fit line is: {rmse_best_fit:.2f}")
print("\nThere is only a slight difference between the estimated line of best-fit and the actual one. This is possible because there is a strong negative linear relationship between vehicle weight and highway miles per gallon.")
print("\nThe actual line of best-fit represents the trend of this data more accurately than the one found through interpolation.")
```
